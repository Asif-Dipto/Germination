## 1. Detection

YOLO11 _detection_ models have no suffix and are the default YOLO11 models, i.e. `yolo11n.pt` and are pretrained on COCO. See [Detection Docs](https://docs.ultralytics.com/tasks/detect/) for full details.


```
# prompt: code to save the /content/runs/detect/train and /content/runs/detect/val in the /content/drive/MyDrive/ML/4th_year_project/20epoch folder in drive

import shutil
import os

# Define source and destination paths
train_source = '/content/runs/detect/train'
val_source = '/content/runs/detect/val'
destination_folder = '/content/drive/MyDrive/ML/4th_year_project/20epoch'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Copy the train folder
train_destination = os.path.join(destination_folder, 'train')
shutil.copytree(train_source, train_destination, dirs_exist_ok=True)


# Copy the val folder
val_destination = os.path.join(destination_folder, 'val')
shutil.copytree(val_source, val_destination, dirs_exist_ok=True)

print(f"Successfully copied '{train_source}' to '{train_destination}'")
print(f"Successfully copied '{val_source}' to '{val_destination}'")
```

```
# prompt: run yolo inference of 2592 x 1944 resolution images with a weight specified by path on  a folder specified by path
import ultralytics
from ultralytics import YOLO
# Replace with your actual paths
weights_path = '/content/drive/MyDrive/asif_atique/50epoch/train/weights/best.pt'
folder_path = '/content/drive/MyDrive/asif_atique/images/culex1.jpg'


# Load the YOLO model
model = YOLO(weights_path)
# Modify class names
model.model.names = {0: 'Aedes', 1: 'Non-Aedes'}

# Run inference on the folder with specified image size
results = model.predict(source=folder_path, save=True, imgsz=640) # save parameter to save results
```
